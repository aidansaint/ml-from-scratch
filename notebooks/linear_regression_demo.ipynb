# Linear Regression

Linear regression is a supervised learning algorithm that models the relationship between one or more input features and a continuous target variable by learning a linear function that minimises the prediction error, typically using Mean Squared Error (MSE) as the loss function.

### Model equation

The hypothesis function is:

$$
\hat{y} = w_1x_1 + w_2x_2 + \dots + w_nx_n + b
$$

Where:
- $\hat{y}$ is the predicted value.
- $x_1, x_2, \dots, x_n$ are the input features.
- $w_1, w_2, \dots, w_n$ are the learned weights.
- $b$ is the bias term (intercept).

### Loss function

To measure prediction error, we typically use the Mean Squared Error (MSE), defined as:

$$
\text{MSE} = \frac{1}{m} \sum_{i=1}^{m} (y_i - \hat{y}_i)^2
$$

Where:
- $m$ is the number of training examples.
- $y_i$ is the target value for the $i^{\text{th}}$ example.
- $\hat{y}_i$ is the predicted value for the $i^{\text{th}}$ example.

## Setup and imports

```python
import matplotlib.pyplot as plt
import numpy as np
import sys

from mlfs.linear_model import LinearRegression

# Allow imports from parent directory.
sys.path.append("..")
```

## Generate data

To illustrate how linear regression works, we'll first generate a synthetic dataset. This allows us to control the true relationship between the input and output variables and evaluate how well the model learns it.

We will generate data using the equation:

$$
y = 2x + 1 + \epsilon
$$

Where:
- $x$ is the input feature drawn uniformly from the range [0, 5]
- $\epsilon \sim \mathcal{N}(0, 1)$ is Gaussian noise.
- $y$ is the target value.

```python
np.random.seed(42) # For reproducibility

# Define linear relationship.
n_samples = 100
true_slope = 2.0
true_intercept = 1.0

# Generate synthetic data.
X = 5 * np.random.rand(n_samples, 1)
noise = np.random.randn(n_samples, 1)
y = true_slope * X + true_intercept + noise

y = y.flatten() # y should be 1D.

# Plot the data.
plt.scatter(X, y, alpha=0.5)
plt.show()
```

## Train the model

Now that we have the input features `X` and the corresponding targets `y`, we can train a linear regression model to learn the best-fitting slope and intercept from the generated data.

```python
# Create and train the model.
model = LinearRegression()
model.fit(X, y)

# Print learned parameters.
print(f"Learned slope: {model.coef_[0]:.2f}")
print(f"Learned intercept: {model.intercept_:.2f}")
```

## Plot the predictions

Let's visualize the predictions of the trained model by plotting the data and the learned line of best fit.

```python
# Predict values for input features.
y_pred = model.predict(X)

# Plot the data and line.
plt.scatter(X, y, alpha=0.5)
plt.plot(X, y_pred)
plt.show()
```

## Evaluate the model

We use the Mean Squared Error (MSE) to evaluate how well the model fits the data.

```python
# Compute and print Mean Squared Error (MSE).
mse = model.score(X, y)
print(f"Mean Squared Error (MSE): {mse:.4f}")
```

## Conclusion

In this notebook, we built a linear regression model to recover a simple linear relationship from noisy synthetic data. The model accurately estimated the underlying parameters, and the Mean Squared Error (MSE) confirmed a strong fit.

While linear regression is a simple technique, its principles form the foundation for many more advanced machine learning algorithms.
